# The Evolution of Deep Neural Networks in Image Classification:  
## From AlexNet to Inception and Beyond

**Authors:**  
__Patricio Villanueva, Melanie Michelle__

**Abstract:**  
This paper explores major milestones in the application of Convolutional Neural Networks (CNNs) for image classification. Beginning with the revolutionary AlexNet—which dramatically reduced the top-5 error rate on the ImageNet challenge—and continuing with Google's Inception (GoogLeNet) model, the paper discusses key innovations that have shaped modern deep learning. Detailed mathematical explanations of critical operations (convolution, ReLU activation, dropout, etc.) are provided, along with diagrams and graphs illustrating the model architectures.

---

## 1. Introduction

Deep neural networks have transformed computer vision, particularly in tasks like image classification, detection, and segmentation. Recent advances in CNN architectures have surpassed traditional methods by leveraging innovative design strategies and regularization techniques. This work focuses on two pivotal contributions:

- **AlexNet (2012):** A breakthrough model that won the ImageNet competition with a top-5 error rate of 15.3% and popularized the use of GPUs for training deep models.
- **Inception (GoogLeNet, 2014-2015):** Developed by Google, it introduced Inception modules that capture features at multiple scales while optimizing computational resources.

---

## 2. AlexNet: The Breakthrough in Image Classification

### 2.1. Model Overview

AlexNet demonstrated the potential of CNNs on large datasets. Its main contributions include:

- **Deep Architecture:** 8 layers (5 convolutional and 3 fully connected layers).
- **ReLU Activation:** Accelerates training by introducing non-linearity.
- **Dropout:** Reduces overfitting by randomly deactivating neurons during training.
- **Data Augmentation:** Increases data variability through rotations, translations, and reflections.
- **GPU Acceleration:** Enables training with millions of parameters.

### 2.2. Mathematical Explanation

#### Convolution Operation  
$$
S(i, j) = \sum_{m}\sum_{n} I(i+m, j+n) \cdot K(m, n)
$$

where:  
- $I$ is the input image.  
- $K$ is the filter (or kernel).  
- $S(i, j)$ is the output at position $(i,j)$.

#### ReLU Activation Function  
$$
\text{ReLU}(x) = \max(0, x)
$$

This introduces non-linearity into the model and speeds up training by avoiding the vanishing gradient problem.

#### Dropout  
$$
y = \frac{1}{1-p} \cdot x \cdot \text{mask}
$$

where:  
- $x$ is the output of a layer.  
- $\text{mask}$ is a randomly generated binary vector.  
- $p$ is the dropout probability.

### 2.3. Architecture Diagram

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()
capas = ["Input", "Conv1", "ReLU", "Pool1", "Conv2", "ReLU", "Pool2", "FC", "Output"]
G.add_nodes_from(capas)
conexiones = [("Input", "Conv1"), ("Conv1", "ReLU"), ("ReLU", "Pool1"),
              ("Pool1", "Conv2"), ("Conv2", "ReLU"), ("ReLU", "Pool2"),
              ("Pool2", "FC"), ("FC", "Output")]
G.add_edges_from(conexiones)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='skyblue', font_size=10)
plt.title("Basic AlexNet arch")
plt.show()
```


---

## 3. Inception (GoogLeNet): The Google Architecture

### 3.1. Model Overview

The Inception model, developed by Google, introduced modules that perform convolutions of different sizes in parallel, capturing features at multiple scales. Key features include:

- **Inception Modules:** Combining $1 \times 1$, $3 \times 3$, and $5 \times 5$ convolutions with pooling operations in a single layer.
- **Dimensionality Reduction:** Extensive use of $1 \times 1$ convolutions to reduce the number of parameters.
- **Extreme Depth:** Increased network depth without incurring significant computational cost.

### 3.2. Mathematical Explanation of the Inception Module

Within an Inception module, several convolutions are applied in parallel:

- **$1 \times 1$ Convolution:**  
$$
Y_{1\times1} = I * K_{1\times1}
$$  
Used to reduce dimensionality and extract local features.

- **$3 \times 3$ and $5 \times 5$ Convolutions:**  
$$
Y_{3\times3} = I * K_{3\times3} \quad \text{and} \quad Y_{5\times5} = I * K_{5\times5}
$$  
These capture larger-scale patterns.

- **Pooling:**  
A pooling operation (e.g., max pooling) is applied to extract the most significant feature from a region.

The final output of the module is obtained by concatenating these results:

$$
Y_{\text{Inception}} = \text{Concatenate}(Y_{1\times1}, Y_{3\times3}, Y_{5\times5}, \text{Pooling})
$$

### 3.3. Inception Module Diagram

```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo que represente un módulo Inception simplificado
G = nx.DiGraph()
nodos = ["Input", "Conv1x1", "Conv3x3", "Conv5x5", "Pooling", "Concatenar", "Output"]
G.add_nodes_from(nodos)

# Conexiones: a partir de la entrada, se derivan diferentes caminos
G.add_edge("Input", "Conv1x1")
G.add_edge("Input", "Conv3x3")
G.add_edge("Input", "Conv5x5")
G.add_edge("Input", "Pooling")
# Concatenación de salidas
G.add_edge("Conv1x1", "Concatenar")
G.add_edge("Conv3x3", "Concatenar")
G.add_edge("Conv5x5", "Concatenar")
G.add_edge("Pooling", "Concatenar")
G.add_edge("Concatenar", "Output")

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='lightgreen', font_size=10)
plt.title("Simplify Module of Inception")
plt.show()
```

## 4. Detailed Mathematical Explanations and Additional Details

### 4.1. Forward Propagation and Backpropagation

Training these networks relies on forward propagation and backpropagation to update the weights. The loss function $L$ is minimized using gradients computed as:

$$
\frac{\partial L}{\partial W} = \frac{\partial L}{\partial y} \cdot \frac{\partial y}{\partial W}
$$

where $W$ represents the model's parameters. Optimization is typically performed using variants of Stochastic Gradient Descent (SGD) or adaptive methods like Adam.

### 4.2. Regularization and Optimization Techniques

- **Data Augmentation:** Techniques such as rotation, translation, and scaling are used to increase data diversity.
- **Dropout:** Helps prevent overfitting by randomly deactivating neurons during training.
- **Batch Normalization:** Though not used in AlexNet, it has become popular in modern architectures to stabilize and accelerate training.
